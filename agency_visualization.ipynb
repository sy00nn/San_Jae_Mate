```python
import pandas as pd
import numpy as np
import folium

df1=pd.read_csv("[위고도포함]근로복지공단_산재보험 직업재활기관 현황정보_2019.csv", encoding='utf-8')
df1=df1.iloc[1:]
df1
```

```python
# Folium 지도 객체 생성
m1 = folium.Map(location=[37.5066909, 126.9607295], zoom_start=10)

# 사용자 지정 색상 아이콘 추가
for _, row in df1.iterrows():
    popup = folium.Popup(row['기관명'], max_width=200)
    icon_html = f'''
    <div style="color: white; background-color: #033357; border-radius: 50%; width: 24px; height: 24px; text-align: center; line-height: 24px;">
        <i class="fa fa-star"></i>
    </div>
    '''
    icon = folium.DivIcon(html=icon_html)
    folium.Marker(location=[row['Latitude'], row['Longitude']], popup=popup, icon=icon).add_to(m1)
```
